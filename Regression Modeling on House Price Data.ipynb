## Prepare Data for Machine Learning
### We know that the machine learning algorithms are mathematical expression so we need to provide the data to the algorithm in specific format
### Encoding

```python
# step 1: divide the data based their data type
df_cat=df_house_price.select_dtypes(include = ['object'])
df_cat.columns
```

```python
# using dummy encoding
dummy_var=pd.get_dummies(data = df_cat,drop_first = True)
dummy_var=dummy_var.reset_index()
dummy_var2= pd.get_dummies(data= df_cat)
```

```python
dummy_var.head(2)
```

```python
df_num=df_house_price.select_dtypes(include=[np.number])
df_num.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# separate the dependent and independent variables
df_target=df_house_price['Price']
df_num=df_num.drop('Price',axis=1)
```

```python
# initiate the standard scaler
ss=StandardScaler()
```

```python
# transfrom the data
# when we want to punch the data them fit is used and when we want to change the nature of the data then the fit_transform is used
# when we give some data to microprocessor then it is called stamping
scale=ss.fit_transform(df_num)
scale
```

```python
df_scale=pd.DataFrame(scale,columns=df_num.columns)
df_scale
df_scale=df_scale.reset_index()
```

```python
# make a dataframe for independent variable
df_independent=pd.concat([df_scale,dummy_var],axis=1)
df_independent
```

```python
# split the data for training and testing
import statsmodels
import statsmodels.api as sm

#add a constant column to the dataframe
df_independent=sm.add_constant(df_independent)
df_independent
```

```python
# import training testing module
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_independent,df_target,random_state=10,test_size=0.2)
```

```python
# print dimensions of train set
print('X_train',x_train.shape)
print('Y_train',y_train.shape)

#print dimensions of test set
print('X_test',x_test.shape)
print('Y_test',y_test.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# create a regression model
model=LinearRegression()
```

```python
#fit the model to the training data
model.fit(x_train,y_train)
```

```python
#predict the training data
predictions=model.predict(x_test)
```

```python
# calculate mean squared error
mse=mean_squared_error(y_test,predictions)
print("Mean Squared Error:",mse)
```
